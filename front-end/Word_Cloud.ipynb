```python
import pandas as pd
import altair as alt
import numpy as np
import datetime
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
import spacy
from spacy.lang.pt import Portuguese
from collections import Counter
import re
```

```python
df = pd.read_csv("Alexa3Reviews.csv")
```

```python
nlp = Portuguese()
word_count = Counter()

for line in df["text"]: 
  doc = nlp(line)
  for token in doc:
    if (token.is_punct == False and token.is_stop == False):
      word_count[token.text] += 1

word_count = pd.Series(dict(sorted(word_count.items(), key = lambda item: item[1], reverse=True)))[0:51]
```

```python
word_count = list(word_count.index)
```

```python
word_count = ', '.join(str(e) for e in word_count)
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(word_count)
plt.figure(figsize=(20,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
